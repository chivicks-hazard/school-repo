```python
# Test example 1
import matplotlib.pyplot as plt

x = [1, 2, 3, 4, 5]
y = [2, 4, 6, 8, 10]

# Create a basic line plot
plt.plot(x, y)

# Add a title and labels
plt.title("Basic Line Plot")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")

# Display the plot
plt.show()
```

```python
# Test Example 2: Custom Bar Chart
import matplotlib.pyplot as plt

# Data - Make changes to the given data
categories = ["Product A", "Product B", "Product C"]
sales = [150, 200, 170]

# Create a bar chart
plt.bar(categories, sales, color=["red", "blue", "green"], edgecolor="black")

# Add titles and labels
plt.title("Sales Comparison")
plt.xlabel("Products")
plt.ylabel("Sales (in units)")

# Annotate bars with values
bars = plt.bar(categories, sales)
for bar in bars:
    yval = bar.get_height()
    plt.text(bar.get_x() + bar.get_width() / 2, yval, round(yval, 2), ha="center", va="bottom")

# Show the plot
plt.show()
```

```python
# Test Example 3: Custom Histogram
import matplotlib.pyplot as plt

# Data: Sample test scores
test_scores = [50, 52, 55, 60, 62, 65, 67, 70, 72, 75, 78, 80, 82, 85, 88, 90, 95, 97, 100]

# Create a histogram with custom bins, colors, and edge colors
plt.hist(test_scores, bins=6, color="lightblue", edgecolor="black")

# Add titles and labels
plt.title("Distribution of Test Scores")
plt.xlabel("Score Range")
plt.ylabel("Frequency of Students")

# Display the plot
plt.show()
```

```python
# Test example 4: Overlaying a Density Curve
import matplotlib.pyplot as plt
import numpy as np
from scipy.stats import norm

# Data: Randomly generated numbers
data = np.random.randn(1000)

# Plot the histogram with density
plt.hist(data, bins=30, density=True, alpha=0.6, color="g")

# Plot the normal distribution curve
xmin, xmax = plt.xlim()
x = np.linspace(xmin, xmax, 100)
p = norm.pdf(x, 0, 1)
plt.plot(x, p, "k", linewidth=2)

# Add a title
plt.title("Histogram with Normal Distribution Overlay")

# Display the plot
plt.show()
```

```python
# Test Example 5: Custom Scatter Plot
import matplotlib.pyplot as plt

# Data
x = [1, 2, 3, 4, 5, 6, 7, 8]
y = [5, 7, 6, 8, 5, 7, 8, 9]
sizes = [20, 50, 80, 200, 60, 120, 140, 180]
colors = [10, 20, 30, 40, 50, 60, 70, 80]

# Create a scatter plot with custom sizes and colors
plt.scatter(x, y, s=sizes, c=colors, cmap="coolwarm", alpha=0.8, edgecolor="black")

# Add titles and labels
plt.title("Custom Scatter Plot with VAriable Size and Color")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")

# Display the plot
plt.colorbar()
print(plt.show())
```
